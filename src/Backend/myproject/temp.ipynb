```python
import json
f = open('./metadata.json')
data = json.load(f)
data=data["available_items"]
```

```python
import random
# Get a list of all items
all_items = list(data.keys())

# Randomly select half of the items
items_to_set_to_zero = random.sample(all_items, len(all_items) // 2)

# Set the selected items' values to 0
for item in items_to_set_to_zero:
    data[item] = 0

data
```
