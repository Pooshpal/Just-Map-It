4. Make a weighted adjacency Matrix for this network. Reduce weight for main_paths

```python
import math
import numpy as np 
def adjmat(coord_dict,connections):

    dict_len = len(coord_dict) 
    #adj_mat  = [[float('inf')] * dict_len for _ in range(dict_len)]
    adj_mat  = [[0] * dict_len for _ in range(dict_len)]
    for i in connections:
        x                           =       coord_dict[str(i[0])]
        y                           =       coord_dict[str(i[1])]
        dist                        =       int((y[0]-x[0])**2+(y[1]-x[1])**2)
        dist_sqr                    =       int(math.sqrt(dist))
        if [i[0],i[1]] in main_paths:
            dist_sqr = 0.75*dist_sqr
        adj_mat[i[0]-1][i[1]-1]   =       dist_sqr
        adj_mat[i[1]-1][i[0]-1]   =       dist_sqr
    
    return adj_mat
#matrix = adjmat(coord_dict,connections)
```

5.Find the ordered node traversal for shortest path given start and end node

```python
import heapq

def shortest_path(matrix, start, end):
    n = len(matrix)
    distances = [float('inf')] * n
    distances[start] = 0
    visited = [False] * n
    queue = [(0, start)]
    path = {}
    heapq.heapify(queue)

    while queue:
        (dist, node) = heapq.heappop(queue)
        if node == end:
            break
        if not visited[node]:
            visited[node] = True
            for i in range(n):
                if matrix[node][i] != 0:
                    new_dist = dist + matrix[node][i]
                    if new_dist < distances[i]:
                        distances[i] = new_dist
                        path[i] = node
                        heapq.heappush(queue, (new_dist, i))

    shortest_path = [end]
    path_tuples = []
    while end != start:
        shortest_path.append(path[end])
        path_tuples.append((path[end], end))
        end = path[end]
    shortest_path.reverse()
    path_tuples.reverse()

    #print(shortest_path)
    #for [a,b] in path_tuples:
    #    if [a+1,b+1] in main_paths:
    #        print(a+1," - ",b+1," - Main Path")
    #    else:
    #        print(a+1," - ",b+1)
    #print(shortest_path)
    return(path_tuples)

#path = shortest_path(matrix, start = 0, end = 84)
```

```python
import heapq

def shortest_path_3d(matrix, start, end):
    n = len(matrix)  # Number of floors
    distances = [[[float('inf')] * n for _ in range(len(matrix[0]))] for _ in range(len(matrix[0][0]))]
    distances[start[2]][start[0]][start[1]] = 0
    visited = [[[False] * n for _ in range(len(matrix[0]))] for _ in range(len(matrix[0][0]))]
    queue = [(0, start)]
    path = {}
    heapq.heapify(queue)

    while queue:
        (dist, (x, y, floor)) = heapq.heappop(queue)
        if (x, y, floor) == end:
            break
        if not visited[floor][x][y]:
            visited[floor][x][y] = True
            for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
                new_x, new_y = x + dx, y + dy
                if 0 <= new_x < len(matrix[0]) and 0 <= new_y < len(matrix[0][0]) and matrix[floor][new_x][new_y] != 0:
                    new_dist = dist + matrix[floor][new_x][new_y]
                    if new_dist < distances[floor][new_x][new_y]:
                        distances[floor][new_x][new_y] = new_dist
                        path[(new_x, new_y, floor)] = (x, y, floor)
                        heapq.heappush(queue, (new_dist, (new_x, new_y, floor)))

            for df in [-1, 1]:
                new_floor = floor + df
                if 0 <= new_floor < n and matrix[new_floor][x][y] != 0:
                    new_dist = dist + matrix[new_floor][x][y]
                    if new_dist < distances[new_floor][x][y]:
                        distances[new_floor][x][y] = new_dist
                        path[(x, y, new_floor)] = (x, y, floor)
                        heapq.heappush(queue, (new_dist, (x, y, new_floor)))

    shortest_path = [(end[0], end[1], end[2])]
    while (end[0], end[1], end[2]) != (start[0], start[1], start[2]):
        shortest_path.append(path[(end[0], end[1], end[2])])
        end = path[(end[0], end[1], end[2])]

    shortest_path.reverse()
    return shortest_path

# Example usage:
matrix = [
    [
        [1, 2],
        [0, 3]
    ],
    [
        [0, 0],
        [4, 5]
    ]
]

start = (0, 0, 0)  # Starting point (x, y, floor)
end = (1, 1, 1)    # Ending point (x, y, floor)

path = shortest_path_3d(matrix, start, end)
print(path)
```

```python
def generate_path(access_points,start,end):
    matrix = adjmat(coord_dict,connections)
    path = set()
    access_points.sort()
    for i in range(len(access_points)-1):
        temp_path = shortest_path(matrix,access_points[i]-1,access_points[i+1]-1)
        for i in temp_path:
            path.add(i)
        #print(i)
    temp_path_start = shortest_path(matrix,start,access_points[0]-1) 
    temp_path_end = shortest_path(matrix,access_points[-1]-1,end) 
    for i in temp_path_start + temp_path_end:
            path.add(i)
    #print("end",temp_path_end)
    return path
```
